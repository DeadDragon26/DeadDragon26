```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import  StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score

data=pd.read_csv("/content/student_clustering.csv")

x = data[["cgpa"]]
y = data["iq"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)


scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)

model = Perceptron()
model.fit(x_train_scaled, y_train)

y_pred = model.predict(x_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
weight = model.coef_[0]
bias = model.intercept_[0]
print("Weight:", weight)
print("Bias:", bias)
print("Equation y = {}x1 + {}".format(weight[0], bias))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import  StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score


num_students = 100

cgpa = np.random.uniform(1, 10, num_students)

iq = np.random.randint(70, 150, num_students)

package = cgpa * 0.5 + iq * 0.3 + np.random.rand(100,1)*5

if np.std(cgpa) != 0:
    cgpa = (cgpa - np.mean(cgpa)) / np.std(cgpa)
else:
    print("Standard deviation is zero, cannot normalize.")
    cgpa = np.zeros_like(cgpa)

print(cgpa)
iq= (iq-np.mean(iq))/np.std(iq)
package= (package-np.mean(package))/np.std(package)

w1 , w2 , b = np.random.rand(3)
learning_rate =0.01
epochs = 1000

for i in range(epochs):

  y_pred = w1*cgpa + w2*iq + b
  loss = np.mean((y_pred - package)**2)
  dw1 = np.mean((y_pred - package)*cgpa)
  dw2 = np.mean((y_pred - package)*iq)
  db = np.mean(y_pred - package)

  w1 = w1 - learning_rate*dw1
  w2 = w2 - learning_rate*dw2
  b = b - learning_rate*db

  if epochs % 100 == 0:
    print("epoch",i,"loss",loss)

print("weight1",w1)
print("weight2",w2)
print("bias",b)
```
